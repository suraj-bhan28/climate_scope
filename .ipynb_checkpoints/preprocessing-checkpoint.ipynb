```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df=pd.read_csv("global_weather_clean.csv")
```

```python
df.sample(5)
```

```python
irrelevant_cols = [
    'temperature_fahrenheit', 'feels_like_fahrenheit',
    'wind_mph', 'gust_mph', 'gust_kph',
    'pressure_in', 'precip_in', 'visibility_miles',
    'condition_text', 'last_updated_epoch',
    'sunrise', 'sunset', 'moonrise', 'moonset',
    'moon_phase', 'moon_illumination'
]
```

```python
df = df.drop(columns=irrelevant_cols)
```

```python
df.head()
```

```python
df.info()
```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'], format='%d-%m-%Y %H:%M')
```

```python
df['country'] = df['country'].astype('category')
df['location_name'] = df['location_name'].astype('category')
df['timezone'] = df['timezone'].astype('category')
df['wind_direction'] = df['wind_direction'].astype('category')
```

```python
df['humidity'] = df['humidity'].astype('int32')
```

```python
df['temperature_celsius'] = df['temperature_celsius'].astype('float32')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python

# Extract month from datetime
df['month'] = df['last_updated'].dt.month

# Compute monthly averages for key weather metrics
monthly_avg = df.groupby('month')[['temperature_celsius','precip_mm','humidity','wind_kph']].mean()

print(monthly_avg)
```

```python
# Group by country and compute averages

regional_avg = df.groupby('country', observed=False)[['temperature_celsius','precip_mm','humidity','wind_kph']].mean()
print(regional_avg.head())
```

```python
df.to_csv("global_weather_clean.csv", index=False)
```
